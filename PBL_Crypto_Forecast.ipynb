```python
split=0.8*df.shape[0]
df_cols=df2.drop('Date', axis=1)
train=df_cols.values[:int(split)]
test=df_cols.values[int(split):]
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler(feature_range=(0,1))

df_train=m.fit_transform(train)
df_test=m.fit_transform(test)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
step=1
window=45
index_col=3

def get(series, index_col, win=45, step=1):
  x=[]
  y=[]

  for j in range(45, len(series)):
    x.append(series[j-win: j:step])
    y.append(series[j,index_col])
  return (np.array(x), np.array(y)) 
x_train, y_train=get(df_train,index_col,window, step)
x_test, y_test=get(df_test,index_col, window, step)
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

# **Deep Learning Model:**

# **LSTM Model:**

```python
import keras
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
model1=Sequential()
model1.add(LSTM(units = 128, return_sequences = True, input_shape=x_train.shape[1:]))
model1.add(Dropout(0.2))
model1.add(LSTM(units = 32, return_sequences = True))
model1.add(Dropout(0.2))
model1.add(LSTM(units = 8, return_sequences = True))
model1.add(Dropout(0.2))
model1.add(LSTM(units = 1))
model1.summary()
```

```python
model1.compile(optimizer='adam', loss='mean_squared_error')
history=model1.fit(x_train, y_train, epochs=128, batch_size=60, validation_data=(x_test, y_test))
```

```python
plt.plot(range(len(history.history['loss'])),history.history['loss'], label='Loss')
plt.plot(range(len(history.history['val_loss'])),history.history['val_loss'], label='Value_Loss')
plt.legend()
```

```python
pred=model1.predict(x_test)
```
